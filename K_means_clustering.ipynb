```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Sample dataset: Annual income and spending score
data = {
    'Income': [15, 16, 17, 18, 25, 60, 62, 63, 64, 100],
    'SpendingScore': [39, 81, 6, 77, 40, 88, 25, 70, 45, 50]
}
df = pd.DataFrame(data)
```

```python
# Feature matrix
X = df[['Income', 'SpendingScore']]

# Fit K-Means with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
# Plotting the clusters
plt.figure(figsize=(8, 5))
plt.scatter(df['Income'], df['SpendingScore'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation with K-Means')
plt.show()
```
